```python
import pandas as pd
import numpy as np

data=pd.read_csv("IPcountry_CCcountry_mismatch_0107-0608.csv")
print data.head(5)
print data['cc_country'][0]
print data.iloc[1], type(data.iloc[1])
print data[1:3]
```

```python
from collections import defaultdict
def cc_ll_pattern_risk(data):
    group={}
    patterns=[]
    columnsd=data.columns
    print columnsd
    for i in range(len(data)):
        #print i
        #print data[i]
        #print data['ll_country'][i]
        pattern=data['cc_country'][i]+'_'+str(data['ll_country'][i])
        if pattern not in patterns:
            patterns.append(pattern)
        if pattern not in group:
            group[pattern]=pd.DataFrame(columns=columnsd)
        #print len(group[pattern])
        group[pattern].loc[len(group[pattern])]=data.iloc[i]
    #print group[pattern]
    risks_list=[]
    Count=[]
    for pattern in patterns:
        table=group[pattern]
        #print pattern
        # print table, type(table)
        
        risk=float(pd.Series.count(table['id']))/float(len(table))
        table['cc_ll_risk']=risk
        group[pattern]=table
        risks_list.append(risk)
        Count.append(len(table))
    risks={ 'risks':risks_list, 'count':Count}
    print risks
    risks=pd.DataFrame(risks,index=patterns)
    return group, risks
    


group,risks=cc_ll_pattern_risk(data) 
print group['US_PE']
#['cc_ll_risk']
#print risks
#risks=pd.Series(risks)
print risks
```

```python
def cc_ll_email_pattern_risk(data):
    group={}
    patterns=[]
    columnsd=data.columns
    print columnsd
    for i in range(len(data)):
        #print i
        #print data[i]
        #print data['ll_country'][i]
        pattern=data['cc_country'][i]+'_'+str(data['ll_country'][i])+data['email'][i]
        if pattern not in patterns:
            patterns.append(pattern)
        if pattern not in group:
            group[pattern]=pd.DataFrame(columns=columnsd)
        #print len(group[pattern])
        group[pattern].loc[len(group[pattern])]=data.iloc[i]
    #print group[pattern]
    risks_list=[]
    Count=[]
    for pattern in patterns:
        table=group[pattern]
        #print pattern
        # print table, type(table)
        
        risk=float(pd.Series.count(table['id']))/float(len(table))
        table['cc_ll_email_risk']=risk
        group[pattern]=table
        risks_list.append(risk)
        Count.append(len(table))
    risks={ 'risks':risks_list, 'count':Count}
    print risks
    risks=pd.DataFrame(risks,index=patterns)
    return group, risks
```

```python
group,risks_email=cc_ll_email_pattern_risk(data)
risks1=risks_email[risks_email['risks']>0.5]
#print risks1
risks2=risks1[risks1['risks']<1.0]
print risks2
risks2=risks2.sort(['risks','count'],ascending=[0,0])
print risks2
```
