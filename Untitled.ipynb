```python
from collections import defaultdict
def cc_ll_pattern_risk(data):
    group={}
    patterns=[]
    columnsd=data.columns
    print columnsd
    for i in range(len(data)):
        #print i
        #print data[i]
        #print data['ll_country'][i]
        pattern=data['cc_country'][i]+'_'+str(data['ll_country'][i])
        if pattern not in patterns:
            patterns.append(pattern)
        if pattern not in group:
            group[pattern]=pd.DataFrame(columns=columnsd)
        #print len(group[pattern])
        group[pattern].loc[len(group[pattern])]=data.iloc[i]
    #print group[pattern]
    risks={}
    for pattern in patterns:
        table=group[pattern]
        #print pattern
        # print table, type(table)
        
        risk=float(pd.Series.count(table['id']))/float(len(table))
        table['cc_ll_risk']=risk
        group[pattern]=table
        risks[pattern]=risk
    return group, risks
    


group,risks=cc_ll_pattern_risk(data) 
print group['US_PE']
#['cc_ll_risk']
#print risks
risks=pd.Series(risks)
print risks
```
